```python
from utils.GraphGenerator import GraphGenerator
from utils.datasets import generate_points
from utils.RunningUtils import do_boruvka_on_graph
import numpy as np

np.random.seed(56)

points = generate_points('blobs', n_samples=500, n_clusters=3)
```

```python
G = GraphGenerator(points.copy()) \
    .to_graph(gen_pair_wise=True)

do_boruvka_on_graph(G, "Blobs graph")
```

```python
G_line = GraphGenerator(points.copy()) \
    .draw_line((-1, 3.5), (0, -5.5), 0.2) \
    .to_graph(gen_pair_wise=True)

do_boruvka_on_graph(G_line, "Blobs graph with line")
```

```python
G_diag = GraphGenerator(points.copy()) \
    .draw_line((0, 5.5), (8, -7.5), 0.2) \
    .to_graph(gen_pair_wise=True)

do_boruvka_on_graph(G_diag, "Blobs graph with diagonal line")
```

```python
G_noise = GraphGenerator(points.copy()) \
    .add_gaussian_noise(0.1) \
    .to_graph(gen_pair_wise=True)

do_boruvka_on_graph(G_noise, "Blobs graph with noise")
```
