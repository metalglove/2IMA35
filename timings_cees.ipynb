```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Harry Plotter

```python
def get_df(filename):
    df = pd.read_csv(filename)
    df = df[['i', 'timing', 'covered_points']]
    return df

def plot(df):
    y = df['timing'].groupby(df['i']).mean()
    x = [i for i in range(1, len(y) + 1)]
    for i in range(1, 11):
        da = df[df['i'] == i]
        x = da['i']
        y = da['timing']
        plt.boxplot(y, positions=[i],  label=f'i: {i}, points in cover: {da['covered_points'].iloc[0]}')
        plt.title(f'Running time vs circle growth')
        plt.xlabel(f'growth (1.2)^i')
        plt.ylabel('seconds')
    plt.legend()
```

```python
df = get_df('experiment5.csv')
plot(df)
```
