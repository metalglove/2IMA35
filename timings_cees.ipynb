```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Harry Plotter

```python
def get_df(filename):
    df = pd.read_csv(filename)
    df = df[['i', 'timing']]
    return df

def plot(df):
    plt.figure()
    x = df['i']
    y = df['timing']
    plt.plot(x, y)
    plt.title(f'Running time vs circle growth')
    plt.xlabel(f'growth (1.2)^i')
    plt.ylabel('seconds')
```

```python
dfa = get_df('experiment5.csv')
plot(dfa)
plot(dfa)
```
