```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Harry Plotter

```python
def get_df(filename):
    df = pd.read_csv(filename)
    df = df[['i', 'timing']]
    return df

def plot(df):
    y = df['timing'].groupby(df['i']).mean()
    x = [i for i in range(1, len(y) + 1)]

    plt.plot(x, y)
    plt.title(f'Running time vs circle growth')
    plt.xlabel(f'growth (1.2)^i')
    plt.ylabel('seconds')
```

```python
df = get_df('experiment5.csv')
plot(df)
```
