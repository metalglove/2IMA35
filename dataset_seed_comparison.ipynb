```python
from algorithms.Graph import Graph
from utils.datasets import generate_points
from utils.GraphGenerator import GraphGenerator
from utils.GraphVisualizer import GraphVisualizer
from algorithms.BoruvkasAlgorithm import BoruvkasAlgorithm
```

```python
import random
from matplotlib import pyplot as plt
import numpy as np
samples = [500, 1000, 1500]
seeds = [42, 56, 69]
fig = plt.figure(figsize=(16, 16))
axs = fig.subplots(3, 3)

for i in range(len(samples)):
    for j in range(len(seeds)):
        blobs_points = generate_points('blobs', n_samples=samples[i], n_clusters=3, seed=seeds[j])
        G = GraphGenerator(blobs_points.copy(), seed=seeds[j]) \
            .to_graph(gen_pair_wise=True)
        gv = GraphVisualizer(G)

        def plot(x):
            gv.plot_neighborhoods(f"round={x} samples={samples[i]} seed={seeds[j]}", axs[i, j])

        alg = BoruvkasAlgorithm(G, 100, False, plot)
        (L, Vs, Es, NGs) = alg.run()
```
