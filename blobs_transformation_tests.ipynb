```python
from algorithms.BoruvkasAlgorithm import BoruvkasAlgorithm
from algorithms.Graph import Graph
from utils.GraphGenerator import GraphGenerator
from utils.GraphVisualizer import GraphVisualizer
from sklearn.datasets import make_blobs
```

```python
n_samples = 150
V, y = make_blobs(n_samples = n_samples, centers = 3, n_features = 2, random_state = 56)

points = V

G = GraphGenerator(points) \
    .to_graph(gen_pair_wise=True)

G_noise = GraphGenerator(list(points)) \
    .add_gaussian_noise(0.1) \
    .add_clustered_noise('circle', 0.1) \
    .add_point((0, 0)) \
    .draw_line((5, 5), (2.5, 0), 0.2) \
    .to_graph(gen_pair_wise=True)
```

```python
gv = GraphVisualizer(G)
gv.plot_graph('Graph')

gv_noise = GraphVisualizer(G_noise)
gv_noise.plot_graph('Graph with noise')
```

```python
max_iterations = 10
plot_graph = lambda i: gv.plot_graph(f'Graph round {i}')
alg = BoruvkasAlgorithm(G, max_iterations)
_ = alg.run(plot_graph = plot_graph)
```

```python
plot_graph_noise = lambda i: gv_noise.plot_graph(f'Graph with noise round {i}')
alg_noise = BoruvkasAlgorithm(G_noise, max_iterations)
_n = alg_noise.run(plot_graph = plot_graph_noise)
```
